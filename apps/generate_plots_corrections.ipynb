# Baseline - Coherent state

```python
init_state = "coherent"

dims = {"atom": 2, "field": 10}

with open(f"../saved_artifacts/training/results_dict_{init_state}_single_shot.pkl", "rb") as f:
    results_dict = pickle.load(f)

models_dict = results_dict["run_4"]["models"]
loss_dict = results_dict["run_4"]["loss"]

plot_expected_values(
    models_dict,
    tfinal,
    n_time_steps,
    init_state,
    params,
    picture,
    dims,
    is_scaled=False,
    plot_input="expected",
)

plot_states(
    models_dict,
    params,
    tfinal,
    init_state=init_state,
    dims=dims,
    n_time_steps=50,
    picture=picture,
    is_scaled=False,
    plot_input="state",
)

plot_loss_functions(loss_dict, skip_param=500)

plot_learned_param(loss_dict, true_param=1, skip_param=500, picture=picture)

plot_fidelity(
    models_dict,
    params,
    tfinal,
    init_state=init_state,
    dims=dims,
    n_time_steps=200,
    picture=picture,
    is_scaled=False,
    plot_input="state",
)
```
